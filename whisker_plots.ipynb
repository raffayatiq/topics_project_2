```python
import json
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.lines import Line2D
```

```python
def make_plot(abr, buf_len, ax, ylim):
    f = open("trace_set_complete.txt" + "_" + str(buf_len) + "_" + abr + "_result.json")
    data = json.load(f)
    
    std_y_axis =[]
    avg_bandwidth_x_axis = []
    bufflen_arr = []
    
    for trace in data.keys():
        
        datum = data[trace]
        
        buf_ratio = datum["buftime"]/(datum["buftime"] + datum["playtime"])
        
        if (buf_ratio == 0.0):
            continue
        std_y_axis.append(buf_ratio)
        avg_bandwidth_x_axis.append(int(datum["average_bandwidth"]))
#         else:
#             std_y_axis.append(0.000000)
#             avg_bandwidth_x_axis.append(int(datum["average_bandwidth"]))
        
#     print(len(std_y_axis))
#     ax.set_xlim(-0.05, 5250)
    ax.set_ylim(ylim[0], ylim[1])
#     ax.set_xlabel("Average bandwidth (kbps)")
    ax.set_ylabel("Buffering Ratio")
    ax.set_title("Maximum buffer length " + str(buf_len) + "s")
    ax.boxplot(std_y_axis)
```

```python
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2)

fig.set_figwidth(6.4*2/1.0)
fig.set_figheight(4.8*2/1.0)
fig.suptitle("Buffering ratio of sessions with rebuffering for ABR MPC")

make_plot("mpc", 4, ax1, (0.95, 1.17))
make_plot("mpc", 6, ax2, (-0.005, 0.04))
make_plot("mpc", 8, ax3, (-0.005, 0.04))
make_plot("mpc", 10, ax4, (-0.005, 0.04))
```
