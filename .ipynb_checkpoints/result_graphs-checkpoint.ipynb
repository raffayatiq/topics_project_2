```python
import json
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.lines import Line2D
```

```python
def make_plot(abr):
    f = open(abr + "_result.json")
    data = json.load(f)
    
    std_y_axis =[]
    avg_bandwidth_x_axis = []
    bufflen_arr = []
    color_scheme = [[255, 255, 255], [255, 200, 200], [247, 150, 150], [252, 0, 0], [128, 0, 128]]

    for trace in data.keys():
        datum = data[trace]
        std_y_axis.append(int(datum["std_bandwidth"]))
        avg_bandwidth_x_axis.append(int(datum["average_bandwidth"]))
        bufflen_arr.append(int(datum["bufflen"]))
    
    plot_colors = []
    for bufflen in bufflen_arr:
        plot_colors.append(color_scheme[(bufflen//2) - 1])
    
    plot_colors = np.array(plot_colors)
    legend_elements = [Line2D([0], [0], marker='o', color='w', label='6s', markerfacecolor=(247/255.0, 150/255.0, 150/255.0), markersize=8),
                      Line2D([0], [0], marker='o', color='w', label='8s', markerfacecolor=(252/255.0, 0/255.0, 0/255.0), markersize=8),
                      Line2D([0], [0], marker='o', color='w', label='10s', markerfacecolor=(128/255.0, 0/255.0, 128/255.0), markersize=8)]
    
    plt.xlabel("Average bandwidth (bps)")
    plt.ylabel("Standard deviation in bandwidth (bps)")
    plt.title(abr.upper())
    plt.scatter(avg_bandwidth_x_axis, std_y_axis, s = 50, c = plot_colors/255.0)
    plt.legend(handles=legend_elements, bbox_to_anchor=(1, 0.5), title="Buffer Length")
```

```python
make_plot("mpc")
```

```python
make_plot("hyb")
```

```python
make_plot("bola")
```

```python
def make_plot(abr, buf_len, ax):
    f = open("trace_set_complete.txt" + "_" + str(buf_len) + "_" + abr + "_result.json")
    data = json.load(f)
    
    std_y_axis =[]
    avg_bandwidth_x_axis = []
    bufflen_arr = []
    
    for trace in data.keys():
        datum = data[trace]
        
        if datum["buftime"] != 0:
            std_y_axis.append(int(datum["std_bandwidth"]))
            avg_bandwidth_x_axis.append(int(datum["average_bandwidth"]))
    
    print(len(std_y_axis))
    ax.set_xlim(-5500*0.05, 5250)
    ax.set_ylim(-7800*0.05, 7800)
    ax.set_xlabel("Average bandwidth (kbps)")
    ax.set_ylabel("Standard deviation in bandwidth (kbps)")
    ax.set_title("Maximum buffer length " + str(buf_len) + "s")
    ax.scatter(avg_bandwidth_x_axis, std_y_axis)
```

```python
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2)

fig.set_figwidth(6.4*2/1.2)
fig.set_figheight(4.8*2/1.0)
fig.suptitle("Sessions with rebuffering for ABR MPC")

make_plot("mpc", 4, ax1)
make_plot("mpc", 6, ax2)
make_plot("mpc", 8, ax3)
make_plot("mpc", 10, ax4)
```

```python
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2)

fig.set_figwidth(6.4*2/1.2)
fig.set_figheight(4.8*2/1.0)
fig.suptitle("Sessions with rebuffering for ABR HYB")

make_plot("hyb", 4, ax1)
make_plot("hyb", 6, ax2)
make_plot("hyb", 8, ax3)
make_plot("hyb", 10, ax4)
```

```python
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2)

fig.set_figwidth(6.4*2/1.2)
fig.set_figheight(4.8*2/1.0)
fig.suptitle("Sessions with rebuffering for ABR BOLA")

make_plot("bola", 4, ax1)
make_plot("bola", 6, ax2)
make_plot("bola", 8, ax3)
make_plot("bola", 10, ax4)
```

```python
def check_if_hyb_and_bola_same(metric, buf_len):
    f = open("trace_set_complete.txt" + "_" + str(buf_len) + "_" + "hyb" + "_result.json")
    hyb_data = json.load(f)
    
    f = open("trace_set_complete.txt" + "_" + str(buf_len) + "_" + "bola" + "_result.json")
    bola_data = json.load(f)
    
    for trace in hyb_data:
        if (hyb_data[trace][metric] != bola_data[trace][metric]):
            return False
        
    return True

for i in range(4, 12, 2):
    print(check_if_hyb_and_bola_same("change", i))
```
